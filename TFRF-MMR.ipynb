```python
with open("../joddy/training/4.txt", 'r') as text_file:
    berita1 = text_file.read()
print(berita1)
```

# 1. Text Preprocessing

## a. Segmentasi Kalimat

```python
from nltk.tokenize import sent_tokenize

segmen = sent_tokenize(berita1)
segmen
```

```python
len(segmen)
```

## b. Case Folding

```python
import string

sentence = [sentence.lower().strip("").strip() for sentence in segmen]
sentence
```

## c. Tokenizing

```python
import re
from nltk.tokenize import word_tokenize

sentence = [sentence.translate(str.maketrans('', '', string.punctuation)) for sentence in sentence]
sentence = [re.sub(r"\d+", "", sentence) for sentence in sentence]

word_tokens = [word_tokenize(sentence) for sentence in sentence]
word_tokens
```

## d. Filtering

```python
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')

listStopword = set(stopwords.words('indonesian'))
listStopword

filter = []
for sent in word_tokens:
      filtered = [s for s in sent if s not in listStopword]
      filter.append(filtered)
filter
```

```python
join = [' '.join(t) for t in filter]
join
```

## e. Stemming

```python
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory

factory = StemmerFactory()
stemmer = factory.create_stemmer()

stemming = [stemmer.stem(sent) for sent in join]
stemming
```

```python
len(stemming)
```

```python
join = " ".join(stemming)
words = join.split(" ")
len(words)
```

# 2. TF-RF

```python
import pandas as pd
import math

# List dokumen
documents = stemming

# Tokenisasi dokumen
tokens_list = [doc.split() for doc in documents]

# Membuat set kata unik dari seluruh dokumen
all_tokens_set = set()
for tokens in tokens_list:
    all_tokens_set.update(tokens)
all_tokens_list = list(all_tokens_set)

# Menghitung TF dari setiap kata dalam setiap dokumen
tf = {}
for word in all_tokens_list:
    tf[word] = []
    for i, tokens in enumerate(tokens_list):
        count = tokens.count(word)
        tf[word].append(count)

# Menghitung RF dari setiap kata
rf = {}
for word in all_tokens_list:
    doc_count_t = 0  # Jumlah dokumen yang mengandung kata t
    doc_count_not_t = 0  # Jumlah dokumen yang tidak mengandung kata t
    for tokens in tokens_list:
        if word in tokens:
            doc_count_t += 1
        else:
            doc_count_not_t += 1
    b = doc_count_t
    c = doc_count_not_t
    rf[word] = math.log10(2 + b / max(1, c))

# Menghitung TF-RF dari setiap kata dalam setiap dokumen
tf_rf = {}
for word in all_tokens_list:
    tf_rf[word] = []
    for i in range(len(documents)):
        tf_rf[word].append(tf[word][i] * rf[word])

# Menampilkan hasil dalam bentuk tabel pandas
tfrf = pd.DataFrame(tf_rf)
tfrf
```

# 3. Cosine Similarity

```python
import pandas as pd

# membuat dataframe berita
berita = pd.DataFrame(tfrf)

# memindahkan baris pertama ke dataframe baru bernama "query"
query = berita.iloc[[0]].reset_index(drop=True)
berita = berita.drop([0]).reset_index(drop=True)
```

```python
import numpy as np
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity

# tf-idf matrix yang sudah dihitung sebelumnya
tf_rf_matrix = pd.DataFrame(berita)
tf_rf_matrix

# query vector
query_vector = pd.DataFrame(query)

# hitung cosine similarity antara query dan dokumen
query_scores = cosine_similarity(tf_rf_matrix, query_vector)

# simpan hasil similarity_scores dalam list
relevansi_query = query_scores.flatten().tolist()

# tampilkan hasil similarity_scores dalam bentuk list
relevansi_query
```

```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

tf_rf = pd.DataFrame(berita)
tf_rf

# Menghitung cosine similarity antara dokumen
similarity_matrix = cosine_similarity(tf_rf)

# Menampilkan hasil similarity dalam bentuk dataframe
similarity_matrix
```

# 4. Maximum Marginal Relevance

```python
import numpy as np
import pandas as pd

def mmr(sim1, sim2, lambda_val, summary_percentage):
    selected_documents = []  # Dokumen yang telah dipilih sebelumnya
    ranked_list = []  # Daftar dokumen yang diurutkan berdasarkan skor MMR
    
    # Menghitung jumlah dokumen yang harus diambil berdasarkan persentase ringkasan yang diinginkan
    num_documents = len(sim1)
    num_summary_documents = int(num_documents * summary_percentage)
    
    while len(selected_documents) < num_summary_documents:
        best_document = None
        best_mmr_score = -np.inf
        
        for i in range(len(sim1)):
            if i not in selected_documents:
                # Menghitung relevansi (Sim1) dan keragaman (Sim2) untuk dokumen i
                relevance = sim1[i]
                if len(selected_documents) == 0:
                    diversity = 0
                else:
                    diversity = max(sim2[i, j] for j in selected_documents)
                
                # Menghitung skor MMR untuk dokumen i
                mmr_score = lambda_val * relevance - (1 - lambda_val) * diversity
                
                if mmr_score > best_mmr_score:
                    best_mmr_score = mmr_score
                    best_document = i
        
        # Menambahkan dokumen terpilih ke dalam daftar dokumen yang telah dipilih sebelumnya
        selected_documents.append(best_document)
        
        # Menambahkan dokumen terpilih ke dalam daftar ranked_list
        ranked_list.append(best_document)
    
    return ranked_list

# Contoh penggunaan
sim1 = relevansi_query
sim2 = similarity_matrix
lambda_val = 0.7
summary_percentage = 0.3  # Mengambil ringkasan dengan compression rate

ranked_list = mmr(sim1, sim2, lambda_val, summary_percentage)
print(ranked_list)

# Menghitung nilai MMR untuk setiap dokumen dalam ranked_list
selected_documents = []
for document in ranked_list:
    relevance = sim1[document]
    if len(selected_documents) == 0:
        selected_documents = [document]
        diversity = 0
    else:
        diversity = max(sim2[document, j] for j in selected_documents)
    mmr_score = lambda_val * relevance - (1 - lambda_val) * diversity
    print("Document", document, "Relevance:", relevance, "Diversity:", diversity, "MMR Score:", mmr_score)
```

```python
import pandas as pd
import nltk

with open("../joddy/testing/30%/S4.txt", 'r') as text_file:
    berita1 = text_file.read()

segmen = sent_tokenize(berita1)

ring_manual = pd.DataFrame({"manual":segmen})  # Membuat DataFrame dari list segmen
ring_manual = ring_manual.drop([0]).reset_index(drop=True)  # Menghapus baris pertama (indeks 0) dan mereset indeks
ring_manual
```

```python
import pandas as pd

# Data frame A sebagai ringkasan manual
df_A = pd.DataFrame(df)

# Data frame B sebagai ringkasan sistem
df_B = pd.DataFrame(ring_manual)

# Menghitung TP (True Positive)
tp = df_B[df_B['manual'].isin(df_A['kalimat']) == True].shape[0]

# Menghitung FP (False Positive)
fp = df_A[df_A['kalimat'].isin(df_B['manual']) == False].shape[0]

# Menghitung FN (False Negative)
fn = df_B[df_B['manual'].isin(df_A['kalimat']) == False].shape[0]
                                                               
# Menghitung Precision
precision = tp / (tp + fp)

# Menghitung Recall
recall = tp / (tp + fn)

# Menghitung F-measure
f_measure = (2 * precision * recall) / (precision + recall)

print("True Positive (TP):", tp)
print("False Positive (FP):", fp)
print("False Negative (FN):", fn)
print("Precision:", precision)
print("Recall:", recall)
print("F-measure:", f_measure)
```
